```python
import pandas as pd
import numpy as np
```

```python
def extract_index(row):
    size, summ, idd = map(int, row["index"].split("_"))
    return pd.Series({"size": size, "sum": summ, "id": idd})
```

```python
def prepare_data(df):
    df = df.join(df.apply(extract_index, axis=1))
    df.columns = map(str.strip, df.columns)
    return df
```

```python
def get_winner(value):
    if np.abs(value) < 10**-3:
        return "D"
    elif value < 0:
        return "R"
    return "C"

def get_description(df):
    desk = df[["size", "sum", "value"]].groupby(["size", "sum"]).describe()
    desk = desk["value"]
    desk["is_det"] = desk.apply(lambda row: np.sign(row["min"]) == np.sign(row["max"]), axis=1)
    desk["winner"] = desk["mean"].apply(get_winner)
    return desk[["is_det", "winner", "min", "max"]]
```

```python
def process_file(path):
    df = pd.read_csv(path)
    df = prepare_data(df)
    return get_description(df)
```

```python
process_file("one_cycle_dseki_r.csv")
```

```python
process_file("one_cycle_dseki_c.csv")
```

```python
process_file("two_cycle_seki_c.csv")
```
